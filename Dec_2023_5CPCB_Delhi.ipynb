```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.colors as mcolors
import shutil
```

```python
df = pd.read_csv('5CPCB_Dec_Analysis.csv')
df.head()
```

```python
df_transposed = df.transpose()
df_filled = df_transposed.apply(lambda x: x.fillna(x.mean()), axis=0)
df_filled.head()
```

```python
def get_color(value):
    """Return a color based on the PM2.5 value."""
    if value > 400:
        return (1, 0, 0)
    elif value > 300:
        return (1, 0.5, 0)
    elif value > 200:
        return (1, 1, 0)
    else:
        return (0, 1, 0)
```

```python
def animate(i):
    ax.clear()
    ax.set_xlim(0, len(df_filled.columns) - 1)
    ax.set_ylim(0, len(df_filled.index) - 1)
    ax.set_title(f"PM2.5 Data for December {i + 1}")
    ax.set_xticks(range(len(df_filled.columns)))
    ax.set_xticklabels(df_filled.columns, rotation=90)
    ax.set_yticks(range(len(df_filled.index)))
    ax.set_yticklabels(df_filled.index)
    for j, row in df_filled.iterrows():
        color = get_color(row[i])
        ax.scatter(i, j, color=color, s=100)
    return ax,
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ani = FuncAnimation(fig, animate, frames=31, interval=1000, blit=False)
plt.close(fig)
```

```python
ani.save('pm25_animation_31sec.mp4', writer='ffmpeg', fps=1, dpi=300)
```
